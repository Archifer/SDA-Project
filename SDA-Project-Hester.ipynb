```python
our_title = 'title'
our_platform = 'platform'
our_score = 'score'
our_release_y = 'release_year'
our_release_m = 'release_month'
our_release_d = 'release_day'

our_rank = 'Rank'
our_genre = 'Genre'
our_publisher = 'Publisher'
our_NA = 'NA_Sales'
our_EU = 'EU_Sales'
our_JP = 'JP_Sales'
our_other = 'Other_Sales'
our_global = 'Global_Sales'


our_keys = [our_title, our_platform, our_score, our_release_y, our_release_m, our_release_d, our_rank, our_genre, \
            our_publisher, our_NA, our_EU, our_JP, our_other, our_global]
```

```python
def get_frequencies_for_key(key, data):
    freqs = dict()
    for row in data:
        if row[key] not in freqs:
            freqs[row[key]] = 1
        else:
            freqs[row[key]] += 1
    return freqs

def plot_with_dict(dictio, title, xlabel, ylabel="Frequency"):
    plt.bar(range(len(dictio)), list(dictio.values()), align='center')
    plt.xticks(range(len(dictio)), list(dictio.keys()), rotation='vertical')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
    
def get_first_n_from_sorted(d, n):
    res = dict()
    sorted_dict = [(k, d[k]) for k in sorted(d, key=d.get, reverse=True)]
    for i in sorted_dict[:n]:
        res[i[0]] = i[1]
    return res

def sort_dict_by_key(d):
    res = dict()
    sorted_dict = sorted(d)
    for i in sorted_dict:
        res[i] = d[i]
    return res

def convert_dict_keys_to_ints(dictio):
    res = dict()
    for d in dictio:
        res[float(d)] = dictio[d]
    return res
```

```python
freqs_for_pl = get_frequencies_for_key(our_platform, our_data)
plot_with_dict(freqs_for_pl, "Number of games per platform", "Platform")

plt.figure(figsize=(20,15))
freqs_for_pu = get_frequencies_for_key(our_publisher, our_data)
plot_with_dict(freqs_for_pu, "Number of games per publisher", "Publisher")

print("*******\n Below is a top 20 of this plot, since this plot is not readable\n******* ")

top20_pu = get_first_n_from_sorted(freqs_for_pu, 20)
plot_with_dict(top20_pu, "Top20 of games per publisher", "Publisher")

freqs_for_gen = get_frequencies_for_key(our_genre, our_data)
plot_with_dict(freqs_for_gen, "Number of games per genre", "Genre")

freqs_for_year = get_frequencies_for_key(our_release_y, our_data)
sorted_freqs_for_year = sort_dict_by_key(freqs_for_year)
plot_with_dict(sorted_freqs_for_year, "Number of games released per year", "Year")

plt.figure(figsize=(20, 15))
freqs_for_scores = get_frequencies_for_key(our_score, our_data)
freqs_for_scores = convert_dict_keys_to_ints(freqs_for_scores)
sorted_freqs_for_scores = sort_dict_by_key(freqs_for_scores)
plot_with_dict(sorted_freqs_for_scores, "Number of games per score", "Score")
```
