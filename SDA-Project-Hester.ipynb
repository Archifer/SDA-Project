# Show plots of data to get an overview

```python
def get_frequencies_for_key(key, data):
    freqs = dict()
    for row in data:
        if row[key] not in freqs:
            freqs[row[key]] = 1
        else:
            freqs[row[key]] += 1
    return freqs

def plot_with_dict(dictio, title, xlabel, ylabel="Frequency", colour=None):
    plt.bar(range(len(dictio)), list(dictio.values()), align='center', color=colour)
    plt.xticks(range(len(dictio)), list(dictio.keys()), rotation='vertical')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
    
def get_first_n_from_sorted(d, n):
    res = dict()
    sorted_dict = [(k, d[k]) for k in sorted(d, key=d.get, reverse=True)]
    for i in sorted_dict[:n]:
        res[i[0]] = i[1]
    return res

def sort_dict_by_key(d):
    res = dict()
    sorted_dict = sorted(d)
    for i in sorted_dict:
        res[i] = d[i]
    return res

def convert_dict_keys_to_floats(dictio):
    res = dict()
    for d in dictio:
        res[float(d)] = dictio[d]
    return res
```

```python
plt.figure(figsize=(20, 15))
freqs_for_pl = get_frequencies_for_key(our_platform, our_data)
plot_with_dict(freqs_for_pl, "Number of games per platform", "Platform", colour='#78b7b4')

plt.figure(figsize=(20,15))
freqs_for_pu = get_frequencies_for_key(our_publisher, our_data)
plot_with_dict(freqs_for_pu, "Number of games per publisher", "Publisher", colour='#78aab7')

print("*******\n Below is a top 20 of this plot, since this plot is not readable\n******* ")

plt.figure(figsize=(20, 15))
top20_pu = get_first_n_from_sorted(freqs_for_pu, 20)
plot_with_dict(top20_pu, "Top20 of games per publisher", "Publisher", colour='#789bb7')

plt.figure(figsize=(20, 15))
freqs_for_gen = get_frequencies_for_key(our_genre, our_data)
plot_with_dict(freqs_for_gen, "Number of games per genre", "Genre", colour='#788bb7')

plt.figure(figsize=(20, 15))
freqs_for_year = get_frequencies_for_key(our_release_y, our_data)
sorted_freqs_for_year = sort_dict_by_key(freqs_for_year)
plot_with_dict(sorted_freqs_for_year, "Number of games released per year", "Year", colour='#787bb7')

plt.figure(figsize=(20, 15))
freqs_for_scores = get_frequencies_for_key(our_score, our_data)
freqs_for_scores = convert_dict_keys_to_floats(freqs_for_scores)
sorted_freqs_for_scores = sort_dict_by_key(freqs_for_scores)
plot_with_dict(sorted_freqs_for_scores, "Number of games per score", "Score", colour='#8578b7')
```

# Big hits per year per quarter on all platforms

```python




def find_max_saled_game_in_dict_with_key(dictio, key):
    max_val = 0.0
    max_row = None
    for row in dictio:
        if float(row[key]) > max_val:
            max_val = float(row[key])
            max_row = row
    if max_row == None:
        max_row = {our_title: '---------', our_genre: '--'}
    return max_row

quarter_1 = ['1', '2', '3']
quarter_2 = ['4', '5', '6']
quarter_3 = ['7', '8', '9']
quarter_4 = ['10', '11', '12']

all_big_hits = []
q1_big_hits = []
q2_big_hits = []
q3_big_hits = []
q4_big_hits = []


def get_bigHits_for_year(year, data, key):
    quarter1 = []
    quarter2 = []
    quarter3 = []
    quarter4 = []
    for row in data:
        if row[our_release_y] == year:
            if row[our_release_m] in quarter_1:
                quarter1.append(row)
            elif row[our_release_m] in quarter_2:
                quarter2.append(row)
            elif row[our_release_m] in quarter_3:
                quarter3.append(row)
            elif row[our_release_m] in quarter_4:
                quarter4.append(row)

    max1 = find_max_saled_game_in_dict_with_key(quarter1, key)
    max2 = find_max_saled_game_in_dict_with_key(quarter2, key)
    max3 = find_max_saled_game_in_dict_with_key(quarter3, key)
    max4 = find_max_saled_game_in_dict_with_key(quarter4, key)
    
    
    return max1, max2, max3, max4
    
    
years = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', 
        '2010', '2011', '2012', '2013', '2014', '2015', '2016']

for y in years:
    big_hits_y = get_bigHits_for_year(y, our_data, our_global)
    print(y)
    kwrt = 1
    
    q1_big_hits.append(big_hits_y[0])
    q2_big_hits.append(big_hits_y[1])
    q3_big_hits.append(big_hits_y[2])
    q4_big_hits.append(big_hits_y[3])
    
    for bh in range(len(big_hits_y)):
        if big_hits_y[bh][our_title] != '---------':
            all_big_hits.append(big_hits_y[bh])
        print('Kwartaal', kwrt, ' : ', big_hits_y[bh][our_title])
        
        kwrt += 1
    
plt.figure(figsize=(20, 15))
freqs_for_gen_hits = get_frequencies_for_key(our_genre, all_big_hits)
plot_with_dict(freqs_for_gen_hits, "Number of big hit games per genre", "Genre", colour='#b6833f')

plt.figure(figsize=(20, 15))
freqs_for_pu_hits = get_frequencies_for_key(our_publisher, all_big_hits)
plot_with_dict(freqs_for_pu_hits, "Number of big hit games per publisher", "Publisher", colour='#843fb6')

plt.figure(figsize=(20, 15))
top20_pu_hits = get_first_n_from_sorted(freqs_for_pu_hits, 20)
plot_with_dict(top20_pu_hits, "Top20 of big hit games per publisher", "Publisher", colour='#b63f72')

plt.figure(figsize=(20, 15))
freqs_for_scores_hits = get_frequencies_for_key(our_score, all_big_hits)
freqs_for_scores_hits = convert_dict_keys_to_floats(freqs_for_scores_hits)
sorted_freqs_for_scores_hits = sort_dict_by_key(freqs_for_scores_hits)
plot_with_dict(sorted_freqs_for_scores_hits, "Number of big hit games per score", "Score", colour='#b6483f')        
```
